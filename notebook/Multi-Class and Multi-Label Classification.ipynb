## (b) Multi-label Classification

## i. Exact Match and Hamming Loss

### Exact Match
The exact match ratio considers an instance correct only if all associated labels are correctly predicted. Therefore, it would be the number of incorrectly classified label tuple by the total number of samples  

### Hamming Loss
The hamming loss would be ratio of the total number of incorrectly classified labels (considering each labelfrom each sample) by the total number of labels.(number of samples * number of labels)
